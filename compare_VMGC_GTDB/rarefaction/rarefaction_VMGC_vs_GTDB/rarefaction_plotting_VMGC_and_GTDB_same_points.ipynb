```python
def get_tail_slope(df, num_points=5):

    num_genomes = df['num_genomes'].sort_values(ascending=False).unique()
    to_extract = num_genomes[1:num_points+1]
    temp = df[df['num_genomes'].isin(to_extract)]
    model = LinearRegression().fit(temp['num_genomes'].values.reshape(-1, 1), temp['centroid_90'].values)
    return model.coef_[0]  # slope
```

```python
def get_tail_slope(df, num_points=5, coords=[]):

    num_genomes = df['num_genomes'].sort_values(ascending=False).unique()

    if len(coords)>0:
        to_extract = [i for i in num_genomes if i in range(coords[0], coords[1]+1)]

    
    else:
        to_extract = num_genomes[1:num_points+1]

    temp = df[df['num_genomes'].isin(to_extract)]
    model = LinearRegression().fit(temp['num_genomes'].values.reshape(-1, 1), temp['centroid_90'].values)
    return model.coef_[0], to_extract  # slope


slopes = {}
for sp in vmgc_counts['species'].unique():

    if sp not in gtdb_counts['species'].values:
        continue

    vmgc_temp = vmgc_counts[vmgc_counts['species']==sp]
    # vmgc_temp['database'] = 'VMGC'
    gtdb_temp = gtdb_counts[gtdb_counts['species']==sp]
    # gtdb_temp['database'] = 'GTDB'

    # gtdb_slope, gtdb_coords= get_tail_slope_gtdb(gtdb_temp)
    gtdb_slope, gtdb_points = get_tail_slope(gtdb_temp)
    vmgc_slope, vmgc_points = get_tail_slope(vmgc_temp, coords=(min(gtdb_points), max(gtdb_points)))
    slopes[vmgc_to_species[sp]] = [vmgc_slope, gtdb_slope, vmgc_points, gtdb_points]

    

df = pd.DataFrame.from_dict(slopes, columns=['VMGC_slope', 'GTDB_slope', 'VMGC_range', 'GTDB_range'], orient='index')
df['magnitude_diff'] = df['GTDB_slope'] / df['VMGC_slope']
df
```

```python
import pandas as pd

# 1. Ensure target_counts has 'species' as a column, not an index
target_counts = gtdb_max[['species', 'num_genomes']].copy()
target_counts.rename(columns={'num_genomes': 'target_n'}, inplace=True)

results = []

# 2. Iterate and find the closest row in vmgc_counts
for _, row in target_counts.iterrows():
    # Filter for the species
    species_subset = vmgc_counts[vmgc_counts['species'] == row['species']]
    
    if not species_subset.empty:
        # Find the row where the absolute difference is smallest
        diffs = (species_subset['num_genomes'] - row['target_n']).abs()
        closest_row = species_subset.loc[diffs.idxmin()].copy()
        
        # Add the target value to this row for easy comparison later
        closest_row['target_n'] = row['target_n']
        results.append(closest_row)

# 3. Build the final dataframe
vmgc_matched = pd.DataFrame(results)

# 4. Reorder columns for better readability
final_cols = ['species', 'target_n', 'num_genomes', 'centroid_90', 'centroid_99']
# Only select columns that actually exist in the result
vmgc_matched = vmgc_matched[[c for c in final_cols if c in vmgc_matched.columns]]

print(vmgc_matched.to_string(index=False))
```
