

### Commands to run PhyloPhlAn in Docker:
Followed this tutorial, which is a similar example to this case: https://github.com/biobakery/biobakery/wiki/PhyloPhlAn-3-Example-01-S-aureus

  docker run -it \
    -v "$(pwd)/input_genomes:/input_genomes" \
    -v "$(pwd)/phylophlan_outputs:/phylophlan_outputs"  \
    -v "$(pwd)/logs:/logs" \
    -v "$(pwd)/phylophlan_db:/phylophlan_db" \
    senseerp/phylophlan:v1 bash
    
  export PATH="/opt/conda/bin:$PATH"

  #this takes a while and has errors but they eventually resolve
  phylophlan_setup_database   \
  -g s__Lactobacillus_crispatus \
  -o phylophlan_db/     \
  --verbose 2>&1 | tee logs/phylophlan_setup_database.log


  phylophlan_write_config_file \
      -o isolates_config.cfg \
      -d a \
      --force_nucleotides \
      --db_aa diamond \
      --map_aa diamond \
      --map_dna diamond \
      --msa mafft \
      --trim trimal \
      --tree1 fasttree \
      --tree2 raxml


  phylophlan    \
    -i input_genomes -o phylophlan_outputs \
    -d phylophlan_db/s__Lactobacillus_crispatus  \
    --trim greedy    \
    --remove_fragmentary_entries   \
    --min_num_entries 22    \
    -t a     \
    -f isolates_config.cfg   \
    --diversity low    \
    --force_nucleotides    \
    --nproc 4   \
    --accurate \
    --verbose 2>&1 | tee logs/phylophlan_output.log
